```python
dataset=pd.read_csv("Churn_Modelling.csv")
dataset
```

```python
# Create X

x=dataset.iloc[:,3:-1].values
x
```

```python
# Create Y

y=dataset.iloc[:,-1].values
y
```

```python
# Lable Encoding - Gender

from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
x[:,2]=le.fit_transform(x[:,2])
x
```

```python
# One Hot Encoding - Geography

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct=ColumnTransformer(transformers=[('encoder',OneHotEncoder(),[1])],remainder='passthrough')
x=np.array(ct.fit_transform(x))
print(x)
```

```python
# Print X Shape

print(x.shape)
```

```python
display(pd.DataFrame(x))
```

```python
y=dataset.iloc[:,-1].values
y
```

```python
# Train Test Split

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
# Standardization

from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.fit_transform(x_test)
```

```python
x_train
```

```python
x_test
```

```python
# Building the ANN

# Initialize Artificial Neural Network as sequence of Layers

ann=tf.keras.models.Sequential()
```

```python
# Create input Layer and First Hidden Layer

''' Dense for hidden Layers Units = Number of Neurons in First hidden layer, Activatio fenction='relu'-rectifier '''

ann.add(tf.keras.layers.Dense(units=6,activation='relu'))
```

```python
# Create second hidden Layer

ann.add(tf.keras.layers.Dense(units=6,activation='relu'))
```

```python
# Create output layer

ann.add(tf.keras.layers.Dense(units=1,activation='sigmoid'))
```

```python
# Training ANN

# Compiling the ANN

''' Optimizer:'''

ann.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
# Training the Model

'''Default Batch Size=32,epochs= Random Number'''

ann.fit(x_train,y_train,batch_size=32,epochs=100)
```

```python
# Prediction (prediction will be at percentage)

y_pred=ann.predict(x_test)
y_pred
```

```python
# Convert the prediction percentage to actual result

'''Assum 0.5 is the threshold for sigmoid actiavtion'''

y_pred=(y_pred>0.5)
print(np.concatenate((y_pred.reshape(len(y_pred),1),y_test.reshape(len(y_test),1)),1))
```

```python
# Create Confusion Matrix and accuracy score

from sklearn.metrics import confusion_matrix,accuracy_score
cm=confusion_matrix(y_test,y_pred)
print(cm)
print(accuracy_score(y_test,y_pred))
```

```python
# Predict For Future Values

print(int(ann.predict(sc.transform([[1,0,0,600,1,40,3,60000,2,1,1,50000]]))>0.5))
```
